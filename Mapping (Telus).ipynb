```python
# Search helper function

def search(dataset, search, atts, category = None):
    if category != None:
        return dataset[(dataset['sap_product_desc_en'].str.contains(search)) & (dataset['category'].str.contains(category))][atts]
    else:
        return dataset[dataset['sap_product_desc_en'].str.contains(search)][atts]
```

```python
atts_trans = ['item_description','sap_product_desc_en', 'category', 'sub_category', 'vendor', 'brand', 'class', 'colour', 'device_oem']
atts_maps = df.columns.to_list()
```

```python
search(trans, 'SPIGEN', atts_trans)
```

```python
search(df, 'ZTE', atts_maps)
```

```python
search(trans, 'ZTE', atts_trans)
```

```python
skus = pd.read_csv("Unmapped Product (SKU).csv", names = ['SKU', 'item_description', 'count'])
skus.head()
```

```python
skus = skus.drop('count', axis = 1)
```

```python
skus.head()
```

```python
unmapped = pd.read_csv("Unmapped Product.csv")
unmapped = unmapped.rename(columns={'ITEM DESCRIPTION':'item_description'})
unmapped = unmapped.drop('SKU', axis = 1)
unmapped.head()
```

```python
unmapped_items = unmapped['item_description'].values

skus = skus.loc[skus['item_description'].isin(unmapped_items)][['SKU', 'item_description']]
```

```python
mapped_skus = pd.merge(skus, unmapped, how = 'left', on = 'item_description')
```

```python
mapped_skus.to_csv("Mapped Products.csv")
```
